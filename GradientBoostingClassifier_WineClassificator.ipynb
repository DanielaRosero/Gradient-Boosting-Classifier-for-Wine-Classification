```python
import numpy as np
import wandb
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize
```

```python
wine = load_wine()
```

```python
X = wine.data
y = wine.target
```

A través del busting classifier ver cómo se comporta el modelo con diferentes hiperparámetros y elegir el que mejor accurancy tiene.

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=357)
```

```python
hiperparametros = {
    'learning_rates':[0.02, 0.15, 0.25, 0.53],
    'max_depths': [2,3,4,6],
    'n_estimators': [45, 90, 130],
    'loss_funtions': ['deviance'], #mide cuán bien nuestro modelo va prediciendo la clase real
    'subsamples': [0.8, 1.0], #extraciones de muestras usadas para probar los estimadores reales. 0.8 deja un 0.2 a la aleatoriedad.
    'min_samples_splits': [2, 4], #Especifica el mínmo de muestras necesarias para generar cada árbol o nodo
    'min_samples_leafs': [1,2]#El mínimo que debería tener la hoja de cada árbol
}
```

* Creación del bucle anidado.


```python
for lr in hiperparametros['learning_rates']:
    for max_depth in hiperparametros['max_depths']:
        for n_estimator in hiperparametros['n_estimators']:
            for loss_funtion in hiperparametros['loss_funtions']:
                for subsample in hiperparametros['subsamples']:
                    for min_sample_split in hiperparametros['min_samples_splits']:
                        for min_sample_leaf in hiperparametros['min_samples_leafs']:

                            # Wandb
                            experiment_name = f"gbm_lr{lr}_depth{max_depth}_est{n_estimator}_loss{loss_funtion}_subsample{subsample}_minsplit{min_sample_split}_minleaf{min_sample_leaf}"
                            wandb.init(project='vinitoupgrade', name = experiment_name, config= hiperparametros)

                            #Train
                            clf = GradientBoostingClassifier(learning_rate=lr,
                                                             max_depth=max_depth,
                                                             n_estimators=n_estimator,
                                                             loss= loss_funtion,
                                                             subsample= subsample,
                                                             min_samples_split= min_sample_split,
                                                             min_samples_leaf= min_sample_leaf,
                                                             random_state= 357,
                                                             validation_fraction= 0.1,
                                                             n_iter_no_change= 5,
                                                             tol= 0.01)
                            
                            clf.fit(X_train, y_train)

                            # Predicciones
                            y_pred = clf.predict(X_test)
                            y_pred_proba = clf.predict_proba(X_test)

                            #Enviar las métricas a Wandb
                            wandb.config.learning_rate = lr 
                            wandb.config.max_depth = max_depth
                            wandb.config.n_estimator = n_estimator
                            wandb.config.loss_funtion = loss_funtion
                            wandb.config.subsample = subsample
                            wandb.config.min_samples_split = min_sample_split
                            wandb.config.min_samples_leaf = min_sample_leaf

                            #Calculate performance metrics
                            accuracy = accuracy_score(y_test, y_pred)
                            f1_macro = f1_score(y_test, y_pred, average='macro')
                            y_test_bin = label_binarize(y_test, classes=np.unique(y))
                            y_pred_bin = y_pred_proba.reshape(-1,len(np.unique(y)))
                            roc_auc_macro = roc_auc_score(y_test_bin, y_pred_bin, average='macro', multi_class='ovr')

                            #log metrics wandb
                            wandb.log({"accuracy" : accuracy , "f1_macro" : f1_macro, 
                                       "roc_auc_macro": roc_auc_macro, "validation_score": clf.train_score_[-1]})

                            #finish
                            wandb.finish()
```
